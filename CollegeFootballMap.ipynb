```python
origin = (39.50, -95.35)
map1 = folium.Map(location=origin,zoom_start=4,attr='')


folium.TileLayer('openstreetmap', control = False).add_to(map1)
#folium.TileLayer('Mapbox Bright', control = False).add_to(map1)

for j in range(5,18):
    clusters_num = j
    coords = df.as_matrix(columns=['latitude', 'longitude'])
    kmeans = sklearn.cluster.KMeans(n_clusters=clusters_num)
    kmeans.fit(coords)
    df['cluster'] = pd.Series(kmeans.labels_, index=df.index)
    #kmeans.cluster_centers_


    conf = folium.FeatureGroup(name= str(j-1) +' Conferences', overlay=False)


    for i in range(0,clusters_num):
        locations = df[['latitude', 'longitude']].loc[df['cluster'] == i].values.tolist()

        #locations = df[['latitude', 'longitude']][0:10].values.tolist()
        try:
            hull = ConvexHull(locations)
            hullpoints = hull.points[hull.vertices].tolist()
            folium.features.PolygonMarker(locations=hullpoints,color='blue',weight=3,fill_color='grey',fill_opacity=0.2).add_to(conf)


        except:
            pass

    conf.add_to(map1)
    
#logos
for index, row in df.iterrows():
    
    popup_text = """
                {team} //
                {stadium} //
                {city}, {state} //
                Coordinates: ({lat}, {long}) //
                Old Conference: {conf}
                """.format(team=row["team"], stadium=row["stadium"], city=row["city"], state=row["state"], lat='%.5f'%(row['latitude']), long='%.5f'%(row['longitude']), conf=row['conference'])
    
    icon = CustomIcon(icon_image = r"http://maxakaplan.com/logos/"+df.loc[index]['logo'].split('\\')[-1][:-1], icon_size=(30, 30))
    marker = folium.Marker(location=[row['latitude'], row['longitude']], icon=icon, popup=folium.Popup(popup_text, parse_html=True, max_width=200)).add_to(map1)

folium.FeatureGroup(name='Choose Option Above', overlay=False).add_to(map1)

#folium.plugins.FloatImage(image='http://maxakaplan.com/wp-content/uploads/2018/02/coollogo_com-15414279.png', bottom=15, left=15).add_to(map1)

folium.LayerControl(collapsed=False).add_to(map1)

folium.plugins.Fullscreen(force_separate_button=True).add_to(map1)

map1
#map1.save('map.html')
```
